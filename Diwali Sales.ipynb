# EXPLORATORY DATA ANALYSIS

```python
plt.figure(figsize=(6, 5)) 
ax = sns.countplot(x='Gender', data=df, palette=['navy', 'salmon'])

for bars in ax.containers: 
    ax.bar_label(bars)
    
 
```

```python
#Gender wise spending
plt.figure(figsize=(6, 5)) 
df1 = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount' , ascending=False)

sns.barplot(x='Gender', y='Amount', data=df1, palette=['navy', 'salmon'])


plt.title("Gender Wise Spending")
```

```python
#age 

ax = sns.countplot(data=df , x='Age Group' , hue = 'Gender' , palette=['lime', 'navy'])

for bars in ax.containers: 
    ax.bar_label(bars)
      
```

*We can see that the highest purchasers are Females between age group of 26-35* 

```python
#state 
#total number of orders from top 10 states


sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,7)})
sns.barplot(data = sales_state, x= 'State' , y='Orders' )
```

```python
#total amount from top 10 states 

sales_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,7)})
sns.barplot(data = sales_state, x= 'State' , y='Amount')
```

*From the above chart we can see that despite Kerala having more orders , it has less purchasing power compared to Haryana and it is not in this chart* .

```python
#Occupation 

sns.set(rc={'figure.figsize':(18,8)})
ax = sns.countplot(data=df , x='Occupation')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
#Total amount from top 10 sectors
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,7)})
sns.barplot(data = sales_state, x= 'Occupation' , y='Amount')
```

*from the above graph , we can see that the top buyers are from IT sector , Healthcare and Aviation*

```python
#Product Category and their resp. sales 

sns.set(rc={'figure.figsize':(25,15)})
ax = sns.countplot(data=df , x='Product_Category')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
sales_category = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_category, x= 'Product_Category' , y='Amount')
```

*From the above chart , the highest sold products are from Food,Clothing&Apparel and Electronics sector and the highest count of orders from clothing&apparel and then food*.

```python
#Top 10 most sold products 

fig1 , ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar' , color='Green')
```

**CONCLUSION :**

*Married women in the age group of 26-35 yrs from UP , Maharashtra and Karnataka working in IT Sectors , Healthcare and Aviation buy more products from Food , Clothing and Electronics categories.*
